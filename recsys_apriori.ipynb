```python
import pandas as pd
url = 'https://raw.githubusercontent.com/machine-learning-red/recsys/main/dataset_base.csv'
df = pd.read_csv(url)
df.head()
```

```python
apriori = df.copy()
for column in df:
    if column != "ID_Cliente":
      apriori.loc[(df[column] > 0), column] = column
      apriori.loc[(df[column] == 0), column] = None
apriori.head()
```
